```python
from queue import Queue

class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

# Binary Tree Creation using Recursion
def BinaryTreeCreation(root):
    print("Enter Data:")
    input_val = int(input().strip())
    if input_val == -1:
        return None
    if root == None:
        root = Node(input_val)
    print("Enter Data Left to", root.data)
    root.left = BinaryTreeCreation(root.left)
    print("Enter Data Right to", root.data)
    root.right = BinaryTreeCreation(root.right)
    return root

# Binary Tree Level Order Traversal
def LevelOrderTraversal(root):
    if root == None:
        return
    q = Queue()
    q.put(root)
    q.put(None)
    while not q.empty():
        temp = q.get()
        if temp == None:
            print()
            if not q.empty():
                q.put(None)
        else:
            print(temp.data, end=" ")
            if temp.left:
                q.put(temp.left)
            if temp.right:
                q.put(temp.right)

root = None
root = BinaryTreeCreation(root)
print("Level Order Traversal:")
LevelOrderTraversal(root)
```
